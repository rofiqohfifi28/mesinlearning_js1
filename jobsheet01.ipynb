```python
# manual normalization

def norm_manual(data):
    n_max = max(data)
    n_min = min(data)
    l_data = len(data) #panjang data

    for i in range(0,l_data):
        data[i] = (data[i] - n_min) / (n_max - n_min)

    return data
```

```python
nilai = [10,13,14,12]
nilai_norm = norm_manual(nilai)
nilai_norm
```

```python
import numpy as np
from sklearn.preprocessing import MinMaxScaler, StandardScaler
```

```python
data = [[100, 0.001],
[8, 0.05],
[50, 0.005],
[88, 0.07],
[4, 0.1]]
data = np.asarray(data)
data.shape

scaler = MinMaxScaler()

scaled = scaler.fit_transform(data)
scaled
```

```python
def stand_manual(data):
    n_mean = np.mean(data)
    n_std = np.std(data)
    l_data = len(data)

    for i in range(0,l_data):
        data[i] = (data[i] - n_mean) / n_std

    return data
```

```python
nilai = [10,11,12,13,14,18,20]
nilai_std = stand_manual(nilai)
nilai_std
```

```python
from sklearn.feature_extraction import DictVectorizer

dv = DictVectorizer()

dt = [
        {'name' : 'POLINEMA'},
        {'name' : 'PENS'},
        {'name' : 'PNJ'},
        {'name' : 'POLBAN'},
        {'name' : 'POLIJE'}
]

poly_dv = dv.fit_transform(dt)
poly_dv.toarray()
```

```python
from sklearn.preprocessing import OneHotEncoder
dummy_encoding = OneHotEncoder(drop = 'first')

de = [
    ['State Polytechnics of Malang'],
      ['Electronic State Polytechnics of Surabaya'],
      ['State Polytechnics of Jakarta']
]

print(dummy_encoding.fit_transform(de).toarray())
```

```python
#corpus / text
corpus = [
    'the house had a tiny little mouse',
    'the cat saw the mouse',
    'the mouse ran away from the house',
    'the cat finally ate the mouse',
    'the end of the mouse story'
]
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer(stop_words='english')
response = vectorizer.fit_transform(corpus)
print(response)
```
